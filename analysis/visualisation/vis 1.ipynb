```python
#Bar Chart showing the count of passengers in easy class
fig = px.bar(df, x='class', title = 'Bar Chart', labels ={'class':'Class','count':'Count'},
              color ='class', color_discrete_sequence=px.colors.sequential.Sunsetdark)
fig.show()
```

```python
fig = px.bar(df, x='class', title = 'Bar Chart', labels ={'class':'Class','count':'Count'},
              color ='class', color_discrete_sequence=px.colors.sequential.Viridis)
fig.show()
```

```python
class_count = df['class'].value_counts().reset_index()
class_count.columns = ['class' , 'count']
fig = px.bar(class_count, x='class', y='count', labels = {'class':'Class', 'count':'Count',},
             title='passenger class distribution' , color='class' , color_discrete_sequence = px.colors.sequential.Viridis)
fig.show()
```

```python
import plotly.figure_factory as ff
```

```python
scater_matrix = ff.create_scatterplotmatrix(df[['age','fare','pclass']] , diag = 'box',index = 'pclass',
                                             colormap = 'Portland' , title = 'Scatterplot Matrix' , height = 700 , width = 700)
scater_matrix.show()
```

```python
fig = px.sunburst(df, path=['age','fare','pclass'],values = 'pclass' ,title='Sunburst Chart')
fig.show()
```
